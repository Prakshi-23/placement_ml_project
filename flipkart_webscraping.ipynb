```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
products = soup.find_all('div',{'class':'tUxRFH'})
```

```python
len(products)
```

```python
products
```

```python
import pandas as pd

df=pd.DataFrame(columns=['Title','Ratings','Price','Feature_1','Feature_2','Feature_3','Feature_4','Feature_5'])
df
```

```python
for i in range(0,len(products)):
    data_list=[]
    title = (products[i].find('div',class_='KzDlHZ').text)
    data_list.append(title)
    ratings = (products[i].find('div',class_='XQDdHH').text)
    data_list.append(ratings)
    price = (products[i].find('div',class_='Nx9bqj _4b5DiR'))
    data_list.append(price)
    feature_1 = (products[i].find('ul',class_='G4BRas'))
    data_list.append(feature_1)
    feature_2 = (products[i].find('li',class_='J+igdf'))
    data_list.append(feature_2)
    feature_3 = (products[i].find('li',class_='J+igdf'))
    data_list.append(feature_3)
    feature_4 = (products[i].find('li',class_='J+igdf'))
    data_list.append(feature_4)
    feature_5 = (products[i].find('li',class_='J+igdf'))
    data_list.append(feature_5)

    df.loc[len(df)]=data_list
```
